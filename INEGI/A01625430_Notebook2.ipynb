```python
# Importamos librerías:
import pandas as pd
import numpy as np
```

```python
# Leemos el primer dataset:
df1=pd.read_csv('/content/RESAGEBURB_09CSV20.csv')
df1
```

```python
# Leemos el segundo dataset:
df2=pd.read_csv('/content/RESAGEBURB_26CSV20.csv')
df2
```

```python
# Leemos el tercer dataset:
df3=pd.read_csv('/content/RESAGEBURB_30CSV20.csv')
df3
```

```python
# Unimos los 3 datasets en un mismo dataframe:
dfc = [df1, df2, df3]
dfc1= pd.concat(dfc)
dfc1
```

```python
dfc1.info() # Dimension del dataset.
```

```python
# Crear nuevo data frame con las variables deseadas.
dfc2=dfc1[['ENTIDAD','MUN','NOM_MUN','LOC','NOM_LOC','AGEB','MZA','P18YM_PB','P18YM_PB_F','P18YM_PB_M','P_18YMAS', 'P_18YMAS_F','P_18YMAS_M']]
dfc2
```

```python
# Verificar datos nulos:
dfc2.isnull().sum()
```

```python
# Reemplazar las distintas formas del dataframe con el que representa datos
# nulos por np.nan:
dfc2.replace("*",np.nan, inplace=True)
dfc2.replace("N/D", np.nan, inplace=True)
dfc2.replace("<NA>", np.nan, inplace=True)
```

```python
# Volver a verificar datos nulos:
dfc2.isnull().sum()
```

```python
# Eliminar las filas con datos nulos:
dfc3=dfc2.dropna()
dfc3
```

```python
# Verificar que se hayan eliminado bien:
dfc3.isnull().sum()
```

```python
# Ver qué tipo de variable utiliza cada columna:
dfc3.dtypes
```

```python
# Cambiamos las variables object que eran números a float:
dfc3['P18YM_PB'] = dfc3['P18YM_PB'].astype(float)
dfc3['P18YM_PB_F'] = dfc3['P18YM_PB_F'].astype(float)
dfc3['P18YM_PB_M'] = dfc3['P18YM_PB_M'].astype(float)
dfc3['P_18YMAS'] = dfc3['P_18YMAS'].astype(float)
dfc3['P_18YMAS_F'] = dfc3['P_18YMAS_F'].astype(float)
dfc3['P_18YMAS_M'] = dfc3['P_18YMAS_M'].astype(float)
dfc3['LOC'] = dfc3['LOC'].astype(float)
dfc3['MUN'] = dfc3['MUN'].astype(float)
dfc3['MZA'] = dfc3['MZA'].astype(float)
```

```python
# Hacemos un dataframe de las variables cuantitativas y otro de las
# cualitativas:
df_cuan = dfc3[['P18YM_PB','P18YM_PB_F','P18YM_PB_M','P_18YMAS', 'P_18YMAS_F','P_18YMAS_M']]
df_cual = dfc3[['ENTIDAD','MUN','NOM_MUN','LOC','NOM_LOC','AGEB','MZA']]
```

```python
# Calcular cuartiles:
q1 = df_cuan.quantile(0.25)
q2 = df_cuan.quantile(0.5)  # Mediana.
q3 = df_cuan.quantile(0.75)

# Mostrar los cuartiles.
print("Primer cuartil (Q1):\n", q1)
print("\nMediana (Q2):\n", q2)
print("\nTercer cuartil (Q3):\n", q3)
```

```python
# Calcular rango intercuartil (IQR).
iqr = q3 - q1

# Definir límites para detectar outliers.
lower_bound = q1 - 1.5 * iqr
upper_bound = q3 + 1.5 * iqr

# Identificar outliers:
outliers = ((df_cuan < lower_bound) | (df_cuan > upper_bound)).any(axis=1)

# Mostrar filas con outliers:
print("Filas con valores atípicos:\n", dfc3[outliers])
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Crear boxplot:
plt.figure(figsize=(10, 6))
sns.boxplot(data=df_cuan)
plt.title("Boxplot de las variables")
plt.xlabel("Variables")
plt.ylabel("Valores")
plt.show()
```

```python
# Hacer graficos de barra:
df_cuan.loc[:, 'P18YM_PB'] = pd.to_numeric(df_cuan['P18YM_PB'], errors='coerce')
df_cuan.loc[:, 'P18YM_PB_F'] = pd.to_numeric(df_cuan['P18YM_PB_F'], errors='coerce')
df_cuan.loc[:, 'P18YM_PB_M'] = pd.to_numeric(df_cuan['P18YM_PB_M'], errors='coerce')
df_cuan.loc[:, 'P_18YMAS'] = pd.to_numeric(df_cuan['P_18YMAS'], errors='coerce')
df_cuan.loc[:, 'P_18YMAS_F'] = pd.to_numeric(df_cuan['P_18YMAS_F'], errors='coerce')
df_cuan.loc[:, 'P_18YMAS_M'] = pd.to_numeric(df_cuan['P_18YMAS_M'], errors='coerce')

plt.figure(figsize=(12, 8))
df3.hist(bins=20, color='skyblue', edgecolor='black', grid=False)
plt.xticks(rotation=45)
plt.gca().ticklabel_format(axis='both', style='plain')
plt.tight_layout()
plt.show()
```

```python
# Calcular la matriz de correlación:
correlation_matrix = df_cuan.corr()

# Visualizar la matriz de correlación como un mapa de calor:
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Mapa de calor de correlación')
plt.show()
```

```python
# Gráfico de pastel variables cualitativas:
for column in df_cual:
    plt.figure(figsize=(8, 8))
    df_cual[column].value_counts().head(10).plot(kind='pie', autopct='%1.1f%%', startangle=140)
    plt.title(f'Gráfico de pastel de la columna "{column}"')
    plt.ylabel('')
    plt.show()
```

```python
# Eliminar valores duplicados en la columna 'ENTIDAD':
dfc_unique = df_cual.drop_duplicates(subset='ENTIDAD')
for column in df_cual:
    plt.figure(figsize=(10, 6))
    sns.countplot(data=dfc_unique, x=column, order=dfc_unique[column].value_counts().index[:10])
    plt.title(f'Diagrama de barras de la columna "{column}"')
    plt.xlabel(column)
    plt.ylabel('Frecuencia')
    plt.xticks(rotation=45)
    plt.show()
```
